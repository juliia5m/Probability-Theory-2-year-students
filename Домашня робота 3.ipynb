## 1 Задача 1.3.12
Знайти ймовiрнiсть того, що серед трьох цифр, кожна з яких вибрана навмання, буде лише 1, 2, 3 рiзних.

Обчисліть теоретичну (повним перебором) та еміричну (симулюванням $100000$ експериментів) імовірності.

```python
#task1
from itertools import product
import numpy as np
numbers = range(10)
 
def teor_prob(p):
    a = list(product(numbers, repeat = 3))
    b = []
    for x in a:
        if len(set(x)) == p:
           b.append(x)
    return len(b)/10**3

def emperical_prob(p):
    a = []
    b = []
    for i in range(100000):
        l = list(np.random.choice(numbers, size=3))
        a.append(l)
    for x in a:
        if len(set(x)) == p:
            b.append(x)
    return len(b)



sp = [1,2,3]
for i in sp:
    print("Theoretical prob when {} different numbers is {}".format(i, teor_prob(i)))
    res = emperical_prob(i)
    print("Empirical prob when {} different numbers is {}".format(i, res/100000))
```

# 2 Задача 1.3.14
З послiдовностi чисел $1, 2, . . . , n$ вибирають навмання $k$ рiзних чисел.
Яка ймовiрнiсть того, що:
1. кожне з вибраних чисел кратне даному числу $p$;
2. кожне з вибраних чисел кратне хоча б одному з двох взаємно простих чисел $p$ i $q$;
3. серед вибраних чисел є хоча б одне кратне $p$?

Напишіть  відповідні функції для обрахунку теоретичної (повним перебором) та еміричної (симулюванням $100000$ експериментів) імовірностей в залежності від параметрів $n, k, p, q$.
Виведіть результат для
- $n = 25, k = 5, p = 3, q = 4$;
- $n = 25, k = 10, p = 3, q = 4$.

```python
#task_2a
import numpy as np
import itertools


def task_2a(n, k, p):
    a = list(itertools.combinations(range(1, n + 1), k))
    b = []
    bool = False
    for i in range(len(a)):
        for j in range(k):
            if a[i][j] % p == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            b.append(a[i][j])
    #print(len(b))
    #print(len(a))
    return len(b)/(len(a))



def empiric(n, k, p):
    b = []
    for _ in range(100000):
        a = list(np.random.choice(list(range(1, n + 1)), size=k, replace=False))
        bool = False
        for i in a:
            if i % p == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            b.append(i)

    return len(b)/ 100000

#-----------------------------------------------------------------------------
#task_2b
def task_2b(n, k, p,q):
    a = list(itertools.combinations(range(1, n + 1), k))
    b = []
    bool = False
    for i in range(len(a)):
        for j in range(k):
            if a[i][j] % p == 0 or a[i][j] % q == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            b.append(a[i][j])
    return len(b)/(len(a))



def empiric_b(n, k, p,q):
    b = []
    for _ in range(100000):
        a = list(np.random.choice(list(range(1, n + 1)), size=k, replace=False))
        bool = False
        for i in a:
            if i % p == 0 or i % q == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            b.append(i)

    return len(b)/ 10**5

#--------------------------------------------------------------------------------------
#task_2c
def task_2c(n, k, p):
    a = list(itertools.combinations(range(1, n + 1), k))
    b = []
    for i in range(len(a)):
        for j in range(k):
            if a[i][j] % p == 0:
                b.append(a[i][j])
                break
    return len(b)/(len(a))



def empiric_c(n, k, p):
    b = []
    for _ in range(100000):
        a = list(np.random.choice(list(range(1, n + 1)), size=k, replace=False))
        for i in a:
            if i % p == 0:
                b.append(i)
                break

    return len(b) / 10**5




val = [[25, 5, 3, 4], [25, 10, 3, 4]]
print("Task_2a (each of the selected numbers is multiples of the given number p)")
print("-------------------------------------------------------------------------")
for i in val:
    print("Theor probability n={}, k={}, p={} is {}".format(i[0], i[1], i[2], '%.30f' % task_2a(i[0], i[1], i[2])))
    print("Empiric probability is {}".format('%.30f' % empiric(i[0], i[1], i[2])))
    print("-------------------------------------------------------------")

print("Task_2b (each of the chosen numbers is multiplicative of at least one of two mutually prime numbers p i q)")
print("----------------------------------------------------------------------------------------------------------")
for i in val:
    print("Theor prob n={}, k={}, p={}, q={} is {}".format(i[0], i[1], i[2], i[3], '%.30f' % task_2b(i[0], i[1], i[2], i[3])))
    print("Empiric probab {}".format('%.30f' % empiric_b(i[0], i[1], i[2], i[3])))
    print("---- ---- ---- ---- ---- ---- ---- ---- ---- ----")


print("Task_2c(there is at least one multiple p among the selected numbers)")
print("--------------------------------------------------------------------")
for i in val:
    print("Theor prob n={}, k={}, p={} is {}".format(i[0], i[1], i[2], '%.30f' % task_2c(i[0], i[1], i[2])))
    print("Empiric prob is {}".format('%.30f' % empiric_c(i[0], i[1], i[2])))





```

# 3 Задача 1.4.6
В коморi знаходяться $n$ пар черевикiв.
З них випадковим чином вибираються $2k$ черевикiв.
Яка ймовiрнiсть того, що серед вибраних черевикiв:
1. вiдсутнi парнi;
2. є рiвно одна комплектна пара;
3. є двi комплектнi пари?

Напишіть  відповідні функції для обрахунку теоретичної (повним перебором) та еміричної (симулюванням $100000$ експериментів) імовірностей в залежності від параметрів $n, k$.
Виведіть результат для
- $n = 8, k = 4$;
- $n = 13, k = 5$.

```python
#task3(a,b and c)
import itertools
import numpy as np
def condition(comb):
    res = 0
    res1 = 0
    res2 = 0
    lenth = len(comb)
    for i in comb:
        if len(i) == len(set(i)):
            res += 1
        elif len(i)  == len(set(i)) + 1:
            res1 += 1
        elif len(i) == len(set(i)) + 2:
            res2 += 1
    return res / lenth, res1 / lenth, res2 / lenth



val = [[8,4],[13,5]]
for i in val:
    shoes = [x for x in range(1, i[0] + 1)] * 2

    comb = list(itertools.combinations(shoes, r=2 * i[1]))
    zero, one, two = condition(comb)
    
    n = 100000
    lst = []
    for j in range(n):
        a = np.random.choice(shoes, size=2 * i[1], replace=False)
        lst.append(a)
    zero1, one1, two1 = condition(lst)
    print(i)
    print("Theoretical prob that there is no paired boots:", zero)
    print("Empirical prob that there is no paired boots:", zero1)
    print("----------------------------------------------------")
    print("Theoretical prob that there is only one pair:", one)
    print("Empirical prob that there is no paired boots:", one1)
    print("----------------------------------------------------")
    print("Theoretical prob that there are two pairs:", two)
    print("Empirical prob that there is no paired boots:", two1)
    print("----------------------------------------------------")
```

# 4 Задача 1.2.15
Нехай $\Omega = {1, 2, \ldots, 2n}$.
Всiм числам приписанi ймовiрностi, пропорцiйнi логарифмам цих чисел.
Знайти цi ймовiрностi.
Знайти ймовiрнiсть того, що в результатi експерименту з’явиться:
1. парне число;
2. непарне число.

Напишіть функцію для обрахунку еміричної (симулюванням $100000$ експериментів) імовірності в залежності від параметра $n$.
Виведіть результат для
- $n = 10$;
- $n = 25$.

```python
#task_4(a and b)
import numpy as np


def task4_t(n):
    num = list(range(1, 2 * n + 1))
    logs = []
    for i in num:
        a = np.log(i)
        logs.append(a)

    c = (1/sum(logs))
    lst = []
    for i in logs:
        b = i*c
        lst.append(b)

    sum_even = 0
    sum_odd = 0
    for i in lst[1::2]:
        sum_even += i
    for j in lst[::2]:
        sum_odd += j

    return "even:",sum_even,"odd", sum_odd


def task4_emp(n):
    even = 0
    odd = 0
    for i in range(100000):
        num = np.random.randint(0, 2*n + 1)
        if num % 2 == 0:
            even += 1
        else:
            odd += 1
    return "even:",even / 10**5, "odd:", odd / 10**5


for n in [10, 25]:
    print( "n =", n,"Theoretical prob:",task4_t(n))
    print("Empirical prob:",task4_emp(n))
    print("----------------------------------------------------------------------")

```
